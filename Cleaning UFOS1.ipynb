```python
# Import Dependencies
import pandas as pd
import datetime as dt
from datetime import datetime
from datetime import date
import numpy as np
```

```python
# Name of the CSV file
file = 'scrubbed.csv'
#Read the file
df = pd.read_csv(file, encoding="UTF-8")
# Preview of the DataFrame
df.head()
```

```python
# Remove blancks in columns
df.columns = df.columns.str.replace(' ', '')
df.columns
```

```python
#Rename columns
df.rename(columns={'duration(seconds)':'duration_seconds',
            'datetime':'date_ocurrence',
           'duration(hours/min)':'duration_hours_min',
           'dateposted':'date_posted'}
           , inplace=True)
df.head()
```

```python
df.dtypes
```

```python
#Convert list found to lower cases
states=['AL','AK','AZ','AR','AS','CA','CO','CT','DE','DC','FL','GA','GU','HI','ID','IL','IN','IA','KS','KY','LA','ME','MD',
        'MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ','NM','NY','NC','ND','MP','OH','OK','OR','PA','PR','RI','SC','SD',
        'TN','TX','TT','UT','VT','VA','VI','WA','WV','WI','WY']

count=0

for state in states:
    word = state.lower()
    states[count]=word
    count += 1
```

```python
#Create new dataframe usa
usa_df=df.loc[df["state"].isin(states)]
print(type(usa_df)) 
usa_df
```

```python
#Fill the missing data in country with 'us" for the sightnings in the states
usa_df=usa_df.assign(country="us")
usa_df.count()
```
